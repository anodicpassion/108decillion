```python
seq_data = {
    "date": [],
    "open": [],
    "high": [],
    "low": [],
    "close": [],
    "ma5": [],
    "ma15": [],
    "lips": [],
}
seq_data
```

```python
def clean_data(data):
    for i in data.split("\n")[1:][::-1]:
        row = i[1:-1].split('''","''')
        try:
            Date = str(row[0])
            Open = float(row[1])
            High = float(row[2])
            Low = float(row[3])
            Close = float(row[4])
            Ma5 = float(row[8])
            Ma15 = float(row[9])
            Lips = float(row[7])
            if Date not in seq_data["date"]:
                seq_data["date"].append(Date)
                seq_data["open"].append(Open)
                seq_data["high"].append(High)
                seq_data["low"].append(Low)
                seq_data["close"].append(Close)
                seq_data["ma5"].append(Ma5)
                seq_data["ma15"].append(Ma15)
                seq_data["lips"].append(Lips)
        except ValueError as e:
            print(e)
```

```python
with open("data/option_chain/nifty_25350/NIFTY 19th SEP 25350 CE (20240919093100000 _ 20240916123800000).csv") as i:
    clean_data(i.read())
with open("data/option_chain/nifty_25350/NIFTY 19th SEP 25350 CE (20240916134300000 _ 20240912103500000).csv") as i:
    clean_data(i.read())
(data.split("\n")[0][1:-1].split('''","'''))
```
