```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
final_data = {"Date": [], "Open": [], "High": [], "Low": [], "Close": [], "5MA": [], "15MA": [], "Lip": []}
final_data
```

```python
# with open("data/nifty50/NIFTY 50 (20221216135000000 _ 20221117105000000).csv", "r") as f:
#     data = f.read()
```

```python
def add_data_in_final(data):
    global final_data
    for i in data.split("\n")[1:]:
        try:
            i = i[1:-1]
            row = i.split('''","''')
            
            Open = float(row[1])
            High = float(row[2])
            Low = float(row[3])
            Close = float(row[4])
            MA5 = float(row[8])
            MA15 = float(row[9])
            Lip = float(row[7])
            
            final_data["Date"].append(str(row[0]))
            final_data["Open"].append(Open - Open)
            final_data["High"].append(High - Open)
            final_data["Low"].append(Low - Open)
            final_data["Close"].append(Close - Open)
            final_data["5MA"].append(MA5 - Open)
            final_data["15MA"].append(MA15 - Open)
            final_data["Lip"].append(Lip - Open)
        except ValueError:
            print(123456)
```

```python
for directory in os.listdir('data/nifty50/'):
    with open(f"data/nifty50/{directory}", "r") as f:
        add_data_in_final(f.read())
```

```python
final_data
```

```python
X, y = [], []
for indx, _ in enumerate(final_data["Date"][:-1]):
    X.append([
        final_data["High"][indx],
        final_data["Low"][indx],
        final_data["Close"][indx],
        final_data["Lip"][indx],
        final_data["5MA"][indx],
        final_data["15MA"][indx]
    ])
    y.append([
        final_data["5MA"][indx+1]
    ])
X = np.array(X)
y = np.array(y)
```

```python
X, y, X.shape, y.shape
```

```python
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
model = Sequential()

# Input layer
model.add(Dense(64, input_dim=6, activation='relu'))  # 2 input features

# Hidden layers
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))  # Regularization to prevent overfitting
model.add(Dense(64, activation='relu'))

# Output layer (1 output for regression task)
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# model = Sequential()
# model.add(LSTM(128, input_shape=(2, 1), return_sequences=False))
# model.add(Dense(64, activation='relu'))
# model.add(Dense(1))
# model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
# Evaluate the model
loss = model.evaluate(X_test, y_test)
print(f"Test Loss: {loss}")
```

```python
# Make predictions
y_pred = model.predict(X_test)

# Invert scaling to get original values
y_pred_original = scaler_y.inverse_transform(y_pred)
y_test_original = scaler_y.inverse_transform(y_test)
```

```python
# Display some predictions vs actual values
for i in range(50):
    print(f"change {((y_pred_original[i][0] / y_test_original[i][0]) * 100 )}")
    print(f"Predicted: {y_pred_original[i][0]:.4f}, Actual: {y_test_original[i][0]:.4f}")
```
